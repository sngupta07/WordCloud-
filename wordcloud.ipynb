### WordCloud

```python
#import dependencies
import numpy as np
from PIL import Image #use to open the image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator #use to generate the wordcloud

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the data or words
with open('word.txt', 'r') as f:
    word_text= f.read()
    
words= word_text.split(',') #preprocessing before using in WordCloud module
```

```python
#To know more about the WordCloud
?WordCloud
```

```python
#Now this one line of code create a WordCloud
wc_1= WordCloud().generate_from_text(' '.join(words))
```

```python
plt.figure(figsize= (8, 6))
plt.imshow(wc_1, interpolation= 'bilinear')
plt.show()
```

```python
#The above WordCloud is pretty simple
```

```python
#We can use any photo to create a WordCloud as the photo is
#Here, I used
image_used_1= Image.open('Happy-New-Year-Emoji-2019-5.jpg')

plt.figure(figsize= (12, 8))
plt.imshow(image_used_1)
plt.show()
```

```python
#We can't able to use the above photo directly
#First have to covert in the form of array
emoji_mask= np.array(image_used_1) #created an array mask
print(emoji_mask)
```

```python
#Here, I passed the few parameters
wc_2 = WordCloud(mask= emoji_mask, width= 1000, height= 800, max_words= 1000, background_color= '#E0FFFF').generate(' '.join(words))
```

```python
plt.figure(figsize= (12, 8))
plt.imshow(wc_2, interpolation= 'bilinear')
plt.show()
```

```python
#used to remove the stopwords from the texts
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
#Read two different text file containing the words related to data science
with open('data_science_words.txt', 'r', encoding= 'latin-1') as f:
    ds_1= f.read()
```

```python
with open('data_science_quotes.txt', 'r') as f:
    ds_2= f.read()
```

```python
#Now, remove the stop words from the text file
ds_1_words= ds_1.split()
print('Length before removing the stopwords: ', len(ds_1_words))
for word in ds_1_words:
    if word in ENGLISH_STOP_WORDS: #condition if stopword present then remove it
        ds_1_words.remove(word)
    else:
        pass
    
print('Length after stoping stopwords: ', len(ds_1_words))
```

```python
#similarly for the 2nd file
ds_2_words= ds_2.split()
print('Length before removing the stopwords: ', len(ds_2_words))
for word in ds_2_words:
    if word in ENGLISH_STOP_WORDS:
        ds_2_words.remove(word)
    else:
        pass
    
print('Length after removing the stopwords: ', len(ds_2_words))
```

```python
#Created a complete word list
word_list= []
word_list.extend(ds_1_words)
word_list.extend(ds_2_words)
```

```python
len(word_list)
```

```python
wc_4 = WordCloud(mask= emoji_mask, width= 1000, height= 800, max_words= 1000, random_state= 101,
               background_color= '#FFFAFA').generate(' '.join(word_list))
```

```python
plt.figure(figsize= (12, 10))
plt.imshow(wc_4, interpolation= 'bilinear')
plt.show()
```
